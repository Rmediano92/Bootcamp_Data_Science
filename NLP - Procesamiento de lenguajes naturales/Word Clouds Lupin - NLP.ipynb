```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
lupin_novel = open('lupin.txt', 'r',encoding="UTF-8").read()
```

```python
len(lupin_novel)
```

```python
stopwords=STOPWORDS
stopwords
```

```python
lupin_wc = WordCloud(background_color='white', max_words=len(lupin_novel), stopwords=stopwords)

lupin_wc.generate(lupin_novel)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(lupin_wc) #imshow muestra imágenes 
ax.axis('off')
plt.show()
```

```python
stopwords.add('said') 
stopwords.add('Yes')
stopwords.add('will')
stopwords.add('must')
stopwords.add('know')
stopwords.add('project')
stopwords.add('Gutenberg')
stopwords.add('one')
stopwords.add('two')

lupin_wc = WordCloud(background_color='white', max_words=len(lupin_novel), stopwords=stopwords)
lupin_wc.generate(lupin_novel)

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(lupin_wc)
ax.axis('off')
plt.show()
```

```python
lupin_mask = np.array(Image.open('sombrero.png')) 
```

```python
lupin_mask.shape
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.imshow(lupin_mask)
plt.axis('off')
plt.show()
```

```python
lupin_wc = WordCloud(background_color="black", max_words=2000, mask=lupin_mask,
               stopwords=stopwords, contour_width=2, contour_color='white') 
lupin_wc.generate(lupin_novel)


fig, ax = plt.subplots(figsize=(20,20))
ax.imshow(lupin_wc, interpolation='bilinear')
ax.axis('off')
lupin_wc.to_file("imagenlupin.png")
plt.show()
```
