### Hacer un DF de las peliculas de https://www.filmaffinity.com/es/topcat.php?id=new_th_es

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.filmaffinity.com/es/topcat.php?id=new_th_es"

response = requests.get(url)
print(bool(response))
soup = BeautifulSoup(response.text, "html.parser")
```

```python
peliculas = soup.find_all("div", class_ = "top-movie") 
peliculas #Encontrar los datos de div con la clase "top-movie" y meterlos en una lista
```

```python
pelicula1 = peliculas[0].text #mostra primera lista 
pelicula1
```

```python
ranking_peliculas = list()
for i in soup.find_all("div", class_ = "top-movie"):
    posicion = i.find("div", class_= "pos").text
    duracion = i.find("div", class_= "duration").text.split("\n")[1][1:8]
    año = i.find("div", class_= "mc-data").text.split("\n")[1]
    titulo = i.find("div", class_= "mc-right").find("a")["title"]
    director = i.find("div", class_ = "director").find("a")["title"]
    genero = i.find("div", class_ = "synop").find("a")["title"]
    actores = [i.text.replace("intervenciones de:","").replace("(","").replace(")","") for i in i.find_all("div", class_ = "credits")][1:]
    puntuacion = i.find("div", class_= "avg-rating").text
    n_reviews = i.find("div", class_= "rat-count").text[1:]
    ranking_peliculas.append([posicion,titulo,año,duracion,genero,director,actores,puntuacion,n_reviews])
ranking_peliculas
```

```python
tabla_pelis = pd.DataFrame(ranking_peliculas, columns= ["posicion","titulo","año","duracion","genero","director","actores","puntuacion","n_reviews"])
tabla_pelis
```
